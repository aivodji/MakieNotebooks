```julia
using CairoMakie, ColorSchemes, Colors
include("makieTheme5.jl") # don't forget to include the theme. 

x = 0:0.05:1
y = x.^2
fig = Figure(resolution = (600, 400))
ax = Axis(fig, xlabel = "x", ylabel = "y")
linea = lines!(x, y, color = :dodgerblue)
fillB = band!(x, fill(0,length(x)), y; color = (:dodgerblue, 0.1))
leg = Legend(fig, [[linea, fillB]], ["Label"], halign = :left, valign = :top)
fig[1, 1] = ax
fig[1, 1] = leg
#save("./results/FigFilled1.svg", fig, pt_per_unit = 0.7);
#save("./results/FigFilled1.png", fig, px_per_unit = 2)
fig
```

```julia
using CairoMakie, ColorSchemes, Colors
include("makieTheme5.jl") # don't forget to include the theme. 

x = LinRange(-10,10,200)
fg(x,μ,σ) = exp.(.-(x.-μ).^2 ./(2σ^2))./(σ*√(2π))

fig = Figure(resolution = (600, 400))
ax = Axis(fig, xlabel = "x", ylabel = "y")
# filled curve 1
linea1 = lines!(x, fg(x, 0.25, 1.5), color = "#E69F00")
fillB1 = band!( x, fill(0,length(x)), fg(x, 0.25, 1.5); color = ("#E69F00", 0.1))
# filled curve 2
linea2 = lines!(x, fg(x, 2, 1), color = "#56B4E9")
fillB2 = band!( x, fill(0,length(x)), fg(x, 2, 1); color = ("#56B4E9", 0.1))
# filled curve 3
linea3 = lines!(x, fg(x, -1, 2), color = "#009E73")
fillB3 = band!( x, fill(0,length(x)), fg(x, -1, 2); color = ("#009E73", 0.1))
leg = Legend(fig, [[linea1, fillB1], [linea2, fillB2], [linea3, fillB3]], 
    ["μ = 0.25, σ = 1.25", "μ = 2, σ = 1", "μ = -1, σ = 2"], halign = :left, valign = :top)
fig[1, 1] = ax
fig[1, 1] = leg
#save("./results/FigFilled2.svg", fig, pt_per_unit = 0.7);
save("./results/FigFilled2.png", fig, px_per_unit = 2)
fig
```

```julia
using CairoMakie, ColorSchemes, Colors
include("makieTheme5.jl") # don't forget to include the theme. 

using Random
Random.seed!(34)
rline(x; b=1, m = 1) = m*x + b
x = 0:0.05:1
y1 = rline.(x, m = 0.6)

x = 0:0.05:1
y1 = rline.(x, m = 0.6)
σ = 0.1* rand(length(x))
y1up = y1 .+ σ
y1do = y1 .- σ

y2 = rline.(x; b = 1.5, m = -0.7)
y2up = y2 .+ σ
y2do = y2 .- σ

fig = Figure(resolution = (600, 400))
ax = Axis(fig, xlabel = "x", ylabel = "y")
lines!(x, y1, color = :red)
band!( x, y1do, y1up; color = (:red, 0.2))
band!( x, y2do, y2up; color = (:black, 0.2))
fig[1, 1] = ax
#save("./results/FigFilled5.svg", fig, pt_per_unit = 0.7)
save("./results/FigFilled5.png", fig, px_per_unit = 2)
fig
```
